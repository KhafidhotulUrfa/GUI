```python
import tkinter as tk
from tkinter import filedialog 
from tkinter.filedialog import askopenfile
from tkinter import *
from tkinter import ttk
import numpy as np
import pandas as pd


window = tk.Tk()
window.configure(bg='lightblue')
window.title("Khafidhotul Urfa_24050120120011")


font1 =('normal',15, 'bold')
font2 =('normal', 12, 'bold')
font3 =('normal', 9, 'bold')

input_frame= Frame (window, bg='#7FE9DE')
input_frame.place(relx=0.01, rely=0.51, relwidth=0.47, relheight=0.75, anchor='w')

output_frame=Frame(window, bg='#FCF9BE')
output_frame.place(relx=0.99, rely=0.51, relwidth=0.47,relheight=0.75, anchor='e')

Jud1 =Label(window, text = 'UAS KOMSTAT LANJUT', bg='lightblue', fg='black', font=font1)
Jud1.pack()

Jud2 = Label (window, text = 'Regresi Linier', bg='lightblue', fg='black', font=font2)
Jud2.pack()

but1= Button (input_frame, text ='Choose File', command = lambda:upload_file())
but1.place(relx=0.5, rely=0, relwidth=0.15, relheight=0.05, anchor='n')

tree= ttk.Treeview(input_frame)
tree.place(relx=0.5, rely=0.07, relwidth=0.6, relheight=0.3, anchor='n')

but2= Button (input_frame, text ='Proces', command = lambda:proc())
but2.place(relx=0.5, rely=0.42, relwidth=0.2, relheight=0.05, anchor='n')

#Output
stat_des=LabelFrame(input_frame,text="Statistic Descriptive",bg='white')
stat_des.place(relx=0.5, rely=0.50, relwidth=0.9, relheight=0.45, anchor='n')
lab_statx=ttk.Label(stat_des,text="")
lab_statx.place(rely=0.5, relx=0.05, relwidth=0.45, relheight=0.8, anchor='w')
lab_staty=ttk.Label(stat_des,text="")
lab_staty.place(rely=0.5, relx=0.95, relwidth=0.45, relheight=0.8, anchor='e')

model=LabelFrame(output_frame, text='Model Regresi')
model.place(relx=0.5, rely=0.03, relwidth=0.6, relheight=0.13, anchor='n')
lab_mod=ttk.Label(model,text="")
lab_mod.place(rely=0.0, relx=0.0)

rmsel=LabelFrame(output_frame, text='Nilai RMSE')
rmsel.place(relx=0.5, rely=0.19, relwidth=0.6, relheight=0.13,anchor='n')
lab_rmse=ttk.Label(rmsel,text="")
lab_rmse.place(rely=0.0, relx=0.0)

r2l=LabelFrame(output_frame, text='R square')
r2l.place(relx=0.5, rely=0.35, relwidth=0.6, relheight=0.13,anchor='n')
lab_r2=ttk.Label(r2l,text="")
lab_r2.place(rely=0.0, relx=0.0)


but3=Button(output_frame, text ='Exit', command =window.destroy)
but3.place(relx=0.5, rely=0.80, relwidth=0.2, relheight=0.05, anchor='n')



#Server
def upload_file():
    global df
    import_file= filedialog.askopenfilename()
    import_file=r"{}". format(import_file)
    read_file =pd.read_csv(import_file)
    df=DataFrame(read_file)

    tree.delete()
    tree['column']=list(df.columns)
    tree['show']= "headings"
    
    for col in tree ['column']:
        tree.heading(col,text=col)
    df_rows=df.to_numpy().tolist()
    for row in df_rows:
        tree.insert("","end",values=row)

def proc ():
    global df
    x=df['SMRG'].values
    y=df['LQ45'].values
    mean_x=np.mean(x.copy())
    mean_y=np.mean(y.copy())
    n=len(x)
    num=0
    den=0
    for i in range(n):
        num=(x[i]-mean_x)*(y[i]-mean_y)
        den=(x[i]-mean_x)**2
    b1=float(num)/float(den)
    b0=mean_y-(b1*mean_x)
    lab_mod["text"]=(b0,"+",b1,"*X1")
    mod=lab_mod["text"] 
    
    rms=0
    for i in range (n):
        y_pred=b0+b1*x[i]
        rmse=(y[i]-y_pred)**2
    rmse=np.sqrt(rmse/n)
    lab_rmse["text"]=rmse
    rml=lab_rmse["text"]
    
    ss_t=0
    ss_r=0
    n=len(x)
    for i in range(n):
        y_pred=b0+b1*x[i]
        ss_t=(y[i]-mean_y)**2
        ss_r=(y[i]-y_pred)**2
    r2 = 1-(ss_r/ss_t)
    lab_r2["text"]=r2
    r2l=lab_r2["text"]

    x1=df['SMRG']
    y2=df['LQ45']
    x_des=x1.describe()
    y_des=y2.describe()
    lab_statx["text"]=("Statistic Descrptive X",x_des)
    lab_staty["text"]=("Statistic Descrptive Y",y_des)
    

window.mainloop()

```
